```python
import pandas as pd
import json
```

```python
path = '/home/gustavo/data-engineering-fakestore/data/bronze/products/products.json'

with open(path) as f:
    data = json.load(f)
```

```python
df = pd.json_normalize(data)
df.head()
```

```python
# capitalize the first letter of each word in `category`
df['category'] = df['category'].apply(lambda s: ' '.join(w.capitalize() for w in s.split()))
df.head()
```
